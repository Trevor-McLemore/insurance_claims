```python
def string_to_numeric(true_value, x):
    if x == true_value:
        x = 1
    else:
        x = 0
    return x

cols_to_recode = {
    'Building_Painted': 'N',
    'Building_Fenced': 'N',
    'Garden': 'V',
    'Settlement': 'R',
}

for k,v in cols_to_recode.items():
    df[k] = df[k].apply(lambda x: string_to_numeric(v, x))
    
df.head()
```

### Transform and scale data

```python
dimension_log = np.log(df['Building Dimension'])
df['dimension_log'] = dimension_log
```

```python
#calculate length of occupancy
df['Occupancy_length_until_observation'] = df['YearOfObservation'] -  df['Date_of_Occupancy']
df['log_occupancy_length'] = np.log(df['Occupancy_length_until_observation']+1)
```
